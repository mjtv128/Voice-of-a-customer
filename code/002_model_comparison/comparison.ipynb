# Model comparison
Compared 5 models (Bert, STM, NMF, LDA with tf-idf, LDA without tf-idf) using coherence scores

```python
import pandas as pd
import numpy as np
from matplotlib.pyplot import figure
import matplotlib.pyplot as plt
```

```python
lda = pd.read_csv('../../data/intermediary_data/lda_cohernece.csv')
lda = lda['0'].tolist()
lda
```

```python
tf = pd.read_csv('../../data/intermediary_data/lda_tfi_coherence.csv')
tf = tf['0'].tolist()
tf
```

```python
nmf = pd.read_csv('../../data/intermediary_data/NNMF_coherence_scores.csv')
nmf = nmf['NNMF_Coherence_Score'].tolist()
nmf
```

```python
column_names=["Num topics","Coherence Score"]
stm = pd.read_csv('../../data/intermediary_data/STM_coherence_score.csv', names = column_names)
stm = stm['Coherence Score'].tolist()
stm
```

```python
column_names=["Num topics","Coherence Score"]
bert = pd.read_csv('../../data/intermediary_data/Numoftopics_Bertopic.csv', names = column_names)
bert = bert['Coherence Score'].tolist()
bert
```

```python
X = range(4, 11)
```

```python
fig, ax = plt.subplots(figsize=(15, 8))

ax.plot(X, tf, color='orange', label='LDA w/ tf-idf')
ax.plot(X, lda, color='r', label='LDA')
ax.plot(X, nmf, color='g', label='NMF')
ax.plot(X, bert, color='b', label='BERT')
ax.plot(X, stm, color='pink', label='STM')

plt.xlabel("Number of Topics")
plt.ylabel("Coherence Scores")
plt.title("Comparison between different topic modelling approaches")
plt.legend()

plt.show()

```
